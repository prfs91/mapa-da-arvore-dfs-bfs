## 1. Objetivo da Atividade

Esta atividade consiste na modelagem de uma estrutura de dados do tipo **Árvore**, baseada no conteúdo da **Aula 03 (16/04/2026)**. O foco central é a implementação prática e a análise técnica de dois algoritmos fundamentais de exploração de grafos:

### 1.1 Algoritmos Implementados

1.  **Busca em Profundidade (DFS - Depth-First Search):**
    *   **Estratégia:** Exploração vertical.
    *   **Comportamento:** Prioriza atingir os nós folha (o ponto mais profundo de um ramo) antes de realizar o *backtracking* para explorar novas ramificações.
    *   **Aplicação:** Ideal para encontrar caminhos distantes e explorar possibilidades completas de uma ramificação.

2.  **Busca em Largura (BFS - Breadth-First Search):**
    *   **Estratégia:** Exploração horizontal.
    *   **Comportamento:** Prioriza a visita de todos os nós de um mesmo nível (camada) antes de descer para o próximo nível da hierarquia.
    *   **Aplicação:** Eficiente para encontrar o caminho mais curto em termos de níveis a partir da raiz.
